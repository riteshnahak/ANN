```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/sunnysavita10/deep-learning/main/aps_data.csv')
```

```python
# no nan showing as valaues written as na
## Convert na value to nan
df=df.replace("na",np.nan)
```

```python
# Filling Null values with median
for feature in [feature for feature in df.columns if feature not in ["class"]]:
  df[feature]=df[feature].fillna(df[feature].median())
```

```python
df.isnull().sum().sum()
```

```python
x=df.drop("class",axis=1)
y=df["class"]
```

```python
x.info(show_counts=True)
```

```python
x.info(verbose=True,show_counts=True)
```

```python
y
```

```python
# Encoding categorical data to numeric value
y=pd.get_dummies(y,drop_first=True)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# As we are using gradient descent as optimizer, so to converge fast to local minima reduced the scale
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

## Lets Built ANN Model



```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
classifier=Sequential()
#lets create 170 input layer for 170 feature
classifier.add(Dense(units=10,kernel_initializer='he_uniform',activation="relu",input_dim=170))
#Hidden Layer
classifier.add(Dense(units=10,kernel_initializer='he_uniform',activation="relu"))
#Output layer
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation="sigmoid"))
```

```python
# Compile all the above layers
classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
classifier.summary()
# In gpt3 175 billion parameter
#The number of parameters in a dense layer is determined by the number of units in the layer and the number of inputs to the layer.
```

```python
# Lets fit the train data
model_history=classifier.fit(x_train,y_train,batch_size=10,epochs=20,validation_split=0.25)
#validation split means from train data it is taking 25% of data for validation during training itself
```

```python
model_history.history.keys()
```

```python
model_history.history["accuracy"]
```

```python
# Lets check train vs test accuracy
plt.plot(model_history.history["accuracy"])
plt.plot(model_history.history["val_accuracy"])
```

```python
# U can see till 12.5 epoc acuuracy was good for both(nearer to each other)
```

```python
plt.plot(model_history.history["loss"])
plt.plot(model_history.history["val_loss"])
plt.title("loss comparision")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(["train",'test'],loc='upper left')
plt.show()
```

```python
y_predict=classifier.predict(x_test)
```

```python
from kerastuner.tuners import RandomSearch
```

```python
def built_model(hp):
  model=Sequential()
  model.add(Dense(units=32,kernel_initializer='he_uniform',activation="relu",input_dim=170))
  model.add(Dense(units=32,kernel_initializer='he_uniform',activation="relu"))
  model.add(Dense(1,activation="sigmoid"))
  optimizer=hp.Choice("optimizer",values=["adam","sgd","rmsprop","adadelta"])
  model.compile(optimizer=optimizer,loss="binary_crossentropy",metrics=['accuracy'])
  return model

```

```python
tuner=RandomSearch(built_model,objective="val_accuracy",max_trials=5,directory='project', project_name='aps-fault')
```

```python
tuner.search(x_train,y_train,validation_data=(x_test,y_test),epochs=5)
```

```python
from tensorflow.keras import optimizers
def build_model(hp):
    model = Sequential()
    for i in range(hp.Int('num_layers', 2, 20)):
        model.add(Dense(units=hp.Int('units_' + str(i),
                                            min_value=32,
                                            max_value=512,
                                            step=32),
                               activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(
        optimizer=optimizers.Adam(
            hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
        loss='binary_crossentropy',
        metrics=['accuracy'])
    return model
```

```python
tuner_second=RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='project3',project_name='aps-fault-sensor')
```

```python
tuner_second.search(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```
